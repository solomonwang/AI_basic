```python
#实现一周的记账功能

incomes = []
outputs = []
balances = dict()

a_week = ['日','一','二','三','四','五','六']
for i in a_week:
    day_income = int(input('周'+i+'收入：'))
    incomes.append(day_income)
    
for i in a_week:
    day_outputs = int(input('周'+i+'支出：'))
    outputs.append(day_outputs)
    
print('7天的收入：')
for i in incomes:
    print(i)
print('7天的支出：')
for i in outputs:
    print(i)
    
for i in range(len(a_week)):
    balances[i] = incomes[i] - outputs[i]
    
print('每天收支结余：')
for day,balance in balances.items():
    print('周',day+1,'结余为：',balance)
    
print('最终结余为：',sum(incomes) - sum(outputs))
```

```python
#实现ATM机器
table = """
===========
请选择操作
1. 查询余额
2. 存款
3. 取款
4. 退出
===========

"""

balance = 100
action = 0

while action != 4:
    action = int(input(table))
    if action == 1:
        print('余额为：', balance)
    else:
        #可操作2，3，4
        if action ==2:
            money = int(input('请输入存款金额：'))
            balance += money
        else:
            #可操作3，4
            if action == 3:
                require = int(input('请输入取款金额：'))
                if require > balance:
                    print('余额不足')
                else :
                    balance -= require
                    print('已取出存款', require, '元')
            else:
                #操作4
                print('已退出')
                break
```
