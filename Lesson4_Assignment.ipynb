````python
# you may need import some package
import numpy as np
from random import shuffle as shu
x = ['快递太慢了！','不好吃','特别难吃','要齁死我了','很划算','下次还来','味道很不错！','香']
y = ['差评','差评','差评','差评','好评','好评','好评','好评']

def shuffle(x,y):
    z = list(zip(x,y)) 
    shu(z) 
    x , y = zip(*z) 
    return x,y

# print result for certify
for i,j in zip(x,y):
    print(i,':',j)
````
